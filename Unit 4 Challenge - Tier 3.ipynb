### 1. Sourcing and Loading 

#### 1.1. Importing Libraries

```python
# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

#### 1.2.  Loading the data
Data from [London Datastore](https://data.london.gov.uk/): a free, open-source data-sharing portal for London-oriented datasets. 

```python
# Read Excel
url_LondonHousePrices = "https://data.london.gov.uk/download/uk-house-price-index/70ac0766-8902-4eb5-aab5-01951aaed773/UK%20House%20price%20index.xls"
properties = pd.read_excel(url_LondonHousePrices, sheet_name='Average price', index_col= None)
```

**2.1. Exploring your data** 

```python
properties
```

**2.2 - 2.3 Cleaning the data**

```python
# Change column label of the dates to 'Months'
properties = properties.rename(columns={'Unnamed: 0': "Months"})
properties["Months"][0] = "Codes"

# Remove all columns but the 32 boroughs
properties = properties.iloc[:,:33].drop("City of London",axis=1)

# Transpose properties and reset index
properties_t = properties.transpose().reset_index()

# Set "Months" as the column label
properties_t.columns = properties_t.iloc[0]

# Renames and sets the index to the Boroughs and their Codes
properties_t = properties_t.rename(columns={"Months":"Boroughs"}).drop([0]).set_index(["Boroughs","Codes"]).apply(pd.to_numeric, errors='coerce')

properties_t
```

```python
# Plot the data just to see what we're looking at
plt.rcParams["figure.dpi"] = 200
properties.iloc[1:].set_index("Months").apply(pd.to_numeric, errors='coerce').resample('A').mean().plot(figsize=(12,9),ylim=(0,1500000),title="Average Price per Month")
```

**2.4.Transforming the data**

```python
# Melted properties_t and grouped the index such that you can easily see all the prices per year per Borough
Melted = pd.melt(properties_t.reset_index(),id_vars=["Boroughs","Codes"]).rename(columns={0:"Date",'value':"Price"}).set_index(["Date","Boroughs","Codes"])

Melted = Melted.groupby([pd.Grouper(level='Boroughs'), 
                pd.Grouper(level='Codes'), 
                pd.Grouper(level="Date",freq="A")]
              ).mean()
Melted
```

**2.6. Visualizing the data**

```python
# Put the borough you want to see in the list L and the loop will plot them price v. year

L = ["Hackney","Waltham Forest","Southwark"]
plt.figure(figsize=(10,7),dpi=200)

for i in L:
    d=Melted.loc[i].reset_index()
    plt.plot(d['Date'],d['Price'],label=i)

plt.xlim(d['Date'][0], d['Date'][len(d['Date'])-1])
plt.ylim(0,1500000)
plt.title("Average Price per Year ")
plt.legend()
```

**3. Modeling**

```python
# Create an empty dictionary to put the average prices in 1998 and 2018 as well as the ration between the two
price_ratio = dict()

# Loop through the boroughs and takes the price of each year and calculates the ratio. Puts these values into
# the price_ratio dict()
for i in Melted.index:
    if i[0] not in price_ratio:
        Boroughs = Melted.loc[i[0]].reset_index().set_index("Date")
        Start = float(Boroughs["Price"].loc["1998"])
        End = float(Boroughs["Price"].loc["2018"])
        price_ratio[i[0]] = [Start, End, End/Start]

# Just a little more finagling to make it look pretty
ratio = pd.DataFrame(price_ratio).transpose()
ratio.rename(columns={0:"1998",1:"2018",2:"Ratio"}).sort_values("Ratio",ascending=False).head()
```

### 4. Conclusion

Based on what I found, Hackney had the highest growth over the last two decades, prices growing over 600%!
